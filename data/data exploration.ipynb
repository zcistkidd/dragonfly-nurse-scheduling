```python
import pandas as pd
from sklearn import preprocessing
```

```python
df_cover = pd.read_csv("data/SECTION_COVER.csv")
df_days_off = pd.read_csv("data/SECTION_DAYS_OFF.csv")
df_shift_off = pd.read_csv("data/SECTION_SHIFT_OFF_REQUESTS.csv")
df_shift_on = pd.read_csv("data/SECTION_SHIFT_ON_REQUESTS.csv")
df_shift = pd.read_csv("data/SECTION_SHIFTS.csv")
df_staff = pd.read_csv("data/SECTION_STAFF.csv")
```

```python
df_cover.columns = df_cover.columns.str.replace(' ', '')
df_cover.columns = df_cover.columns.str.replace(' ', '')
df_cover.columns = df_cover.columns.str.replace(' ', '')
df_days_off.columns = df_days_off.columns.str.replace(' ', '')
df_shift_off.columns= df_shift_off.columns.str.replace(' ', '')
df_shift_on.columns = df_shift_on.columns.str.replace(' ', '')
df_shift.columns = df_shift.columns.str.replace(' ', '')
df_staff.columns = df_staff.columns.str.replace(' ', '')
```

```python
le_shift = preprocessing.LabelEncoder()
le_staff = preprocessing.LabelEncoder()
```

```python
df_cover.columns
```

```python
df_cover['ShiftID_num'] = le_shift.fit_transform(df_cover['ShiftID'])
df_cover
```

```python
df_staff['ID_num'] = le_staff.fit_transform(df_staff['ID'])
df_staff['MaxShifts_1'] = df_staff['MaxShifts'].map(lambda x:x.split('|')[0].split('=')[1])
df_staff['MaxShifts_0'] = df_staff['MaxShifts'].map(lambda x:x.split('|')[1].split('=')[1])
df_staff['MaxShifts_2'] = df_staff['MaxShifts'].map(lambda x:x.split('|')[2].split('=')[1])
df_staff
```

```python
df_days_off['EmployeeID_num'] = le_staff.fit_transform(df_days_off['EmployeeID'])
df_days_off
```

```python
df_shift_off['EmployeeID_num']= le_staff.fit_transform(df_shift_off['EmployeeID'])
df_shift_off['ShiftID_num']= le_staff.fit_transform(df_shift_off['ShiftID'])
df_shift_off
```

```python
df_shift_on['EmployeeID_num']= le_staff.fit_transform(df_shift_on['EmployeeID'])
df_shift_on['ShiftID_num']= le_staff.fit_transform(df_shift_on['ShiftID'])
df_shift_on
```

```python
df_shift['PrevShiftID_num']= le_shift.fit_transform(df_shift['PrevShiftID'])
df_shift['CurrentShiftID_num']= le_shift.fit_transform(df_shift['CurrentShiftID'])
df_shift
```
